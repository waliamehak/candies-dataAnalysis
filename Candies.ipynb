#**Prepare, Clean and Normailze the Data** 
After downloading the datasets into my local drive, I perused them all very carefully. The very first thing that I noticed was that the 2014 dataset is not going to be of much use to us as it is not person oriented but rather product oriented data. What I mean to say here is that, we dont get to know about the personal preference of an individual on the different candies. Neither we get to know their origin, age or sex. All these factors are crucial if I am to know the likability of candies based on region, sex, etc. So, I ignored this dataset. \
Now, I cleaned the data by removing the columns from each dataset one by one that would not aid us by answering the questions asked. These columns essentially were the ones that stored a response to questions not related to the individual's preferences on different candies. \
I made sure that each data set just contains the personal information and  their preference on candies and nothing else. I dropped all the columns in preprocessing until I had the results that I wanted. \
After cleaning the data, I noticed that the datasets have different number of columns. This essentially meant that we dont have the data on candies for all the three years, i.e., some candies were officially added to the survey while others were removed in each survey.\
For me the concern was to make sure that when I join the datasets, the information should fall under the right columns. That is, if there exists data for a candy in first data set, if should concatinate under the data for same candy in the other datasets and not make its own new seperate column. \
For this to happen the column names should be the same for the same candy. I noticed that for 2015, and 2016 the columns names were same and not any significant changes were to be made. However for 2017, the column names were not same. So, I also changed the names of the columns for which there existed data in 2015 and 2016. I did that by using string handling function ( split) on it and change the column names in 2017.
